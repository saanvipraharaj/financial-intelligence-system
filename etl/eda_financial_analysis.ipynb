```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
```

```python
import pandas as pd

companies = pd.read_csv("../data/clean/companies.csv")
profit = pd.read_csv("../data/clean/profitandloss.csv")
balance = pd.read_csv("../data/clean/balancesheet.csv")
cashflow = pd.read_csv("../data/clean/cashflow.csv")
```

# Sector Distribution

```python
top_roe = companies.sort_values(
    by='roe_percentage',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_roe['company_name'],
    top_roe['roe_percentage']
)

plt.title("Top 10 Companies by ROE")
plt.xlabel("Company")
plt.ylabel("ROE Percentage")

plt.xticks(rotation=75)

plt.show()
```

# Top ROE Companies

```python
top_roe = companies.sort_values(by='roe_percentage', ascending=False).head(10)

print(top_roe[['company_name', 'roe_percentage']])
```

# Missing Values

```python
missing_values = profit.isnull().sum()

print(missing_values[missing_values > 0])
```

# Correlation Heatmap

```python
numeric_df = profit.select_dtypes(include=np.number)

plt.figure(figsize=(14,10))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Financial Correlation Heatmap")

plt.show()
```

# Profit Trend Analysis

```python
tcs_profit = profit[profit['company_id'] == 'TCS']

plt.figure(figsize=(10,6))

plt.plot(
    tcs_profit['year'],
    tcs_profit['net_profit'],
    marker='o'
)

plt.title("TCS Net Profit Trend")
plt.xlabel("Year")
plt.ylabel("Net Profit")

plt.show()
```

# Cash Flow Analysis

```python
tcs_cash = cashflow[cashflow['company_id'] == 'TCS']

plt.figure(figsize=(10,6))

plt.plot(
    tcs_cash['year'],
    tcs_cash['net_cash_flow'],
    marker='o'
)

plt.title("TCS Cash Flow Trend")
plt.xlabel("Year")
plt.ylabel("Net Cash Flow")

plt.show()
```
